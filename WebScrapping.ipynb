```python
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
field=[]
name=[]
location=[]
salary=[]
experience=[]

for j in range(1,50):
  print(j)
  headers = {
      'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'
  }

  url = 'https://internshala.com/jobs/page-{}'.format(j)
  webpage = requests.get(url, headers=headers).text
  soup= BeautifulSoup(webpage,'lxml')
  company= soup.find_all('div', class_='internship_meta')



  for i in company:
    name.append(i.find('p', class_='company-name').text.strip())
    try:
      field.append(i.find('a', class_= 'job-title-href').text.strip())
    except:
      field.append(np.nan)


    try:
      location.append(i.find('p',class_='location').text.strip())
    except:
      location.append(np.nan)

    try:
      salary.append(i.find_all( 'div',class_='row-1-item')[0].text.strip())
    except:
      salary.append(np.nan)

    try:
      experience.append(i.find_all('div',class_='row-1-item')[1] .text.strip())
    except:
      experience.append(np.nan)



    df=pd.DataFrame({'Sector':field,'Company Name':name,'Location':location,'Salary':salary,'Experience':experience})
```
