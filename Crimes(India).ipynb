```python
import pandas as pd

df = pd.read_excel('ProjectSheetStates1 (1).xlsx')
print(df.head(20))
jk_index = 13
ladakh_index = 17
```

```python
summed_values = df.iloc[jk_index, 1:] + df.iloc[ladakh_index, 1:]

new_row = pd.Series(
    ['JammuKashmir+Ladakh'] + summed_values.tolist(),
    index=df.columns
)
```

```python
df_new = df.drop(index=[jk_index, ladakh_index])
df_new = pd.concat([df_new, pd.DataFrame([new_row])], ignore_index=True)
df_new.to_excel('ProjectSheetStates_updated.xlsx', index=False)
```

```python
print("✅ New row added and original rows removed! File saved as ProjectSheetStates_updated.xlsx")
df_new["TotalCrime"] = df_new[[
    "Murders in 2021",
    "Rape Victims",
    "Theft",
    "Attempt to Commit Murder",
    "Dacoity",
    "Kidnapping & Abduction",
    "Grievous Hurt / Assault (Total Hurt)",
    "Burglary / House Breaking",
    "Cybercrimes (Total Registered)",
    "Arson"
]].sum(axis=1)
```
